```python
#Since the books and users are in very large quantity and contaning books whose rating is provided by very less users and also 
#contaning users who has provided rating to very less books so they will affect our model and reduce our accuracy so by hit and
# trial method we will remove those books whose rating is provided by less than 50 users and will remove those users who have
# provided rating less than 200 to books
```

```python
ratings['user_id'].value_counts()
```

```python
x=ratings['user_id'].value_counts()>200
```

```python
x[x]
```

```python
x[x].shape  #these are the users who have provided reviews of book more than 200 
```

```python
y=x[x].index  # These are the user id of users who have provided reviews of more than 200 books
```

```python
y
```

```python
ratings=ratings[ratings['user_id'].isin(y)]  # will provide those column in which users present in y 
```

```python
ratings.shape
```

```python
ratings.head()
```

```python
ratings_with_books=ratings.merge(books,on='ISBN')
```

```python
ratings_with_books.head()
```

```python
ratings_with_books.rename(columns={'book':'title'},inplace=True)
```

```python
number_rating=ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
number_rating.head()
```

```python
number_rating.rename(columns={'book':'title','rating':'number of ratings'},inplace=True)
```

```python
number_rating
```

```python
final_rating=ratings_with_books.merge(number_rating,on='title')
```

```python
final_rating.head()
```

```python
final_rating=final_rating[final_rating['number of ratings']>=50]
```

```python
final_rating.head()
```

```python
final_rating.shape
```

```python
final_rating.drop_duplicates(['user_id','title'],inplace=True)
```

```python
final_rating.shape
```

```python
book_pivot=final_rating.pivot_table(columns='user_id',index='title',values='rating')
```

```python
book_pivot
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
book_sparse=csr_matrix(book_pivot)
```

```python
type(book_sparse) #so that no calculation is needed for 0 value contaning element
```

```python
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(algorithm='brute')  # brute means that this algo is calculating with all element present there
```

```python
model.fit(book_sparse) #model is trained
```

```python
distances,suggestions=model.kneighbors(book_pivot.iloc[224,:].values.reshape(1,-1),n_neighbors=6) #237 is index of some book
```

```python
distances # top 5 minimum distances of particular book i.e recommended book
```

```python
suggestions #book id of recommended book
```

```python
for i in range(len(suggestions)):
    print(book_pivot.index[suggestions[i]])
```

```python
def recommend_book(book_name):
    book_id=np.where(book_pivot.index==book_name)[0][0]
    distances,suggestions=model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1),n_neighbors=6)
    
    for i in range(len(suggestions)):
            print(book_pivot.index[suggestions[i]])
    
```

```python
recommend_book('Animal Farm')
```
